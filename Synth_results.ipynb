```python

dpd_gb = []

dpd_or = []

totals = [90, 30000,30000,615,165,582,32561,666,303,294,223]

male_tots = [47,11888,13598,377,133,441,21790,355,206,213,123]

j = 0
for i in results:

    total = totals[j]

    fem_acc = i[0][1][3]
    
    male_acc = i[0][1][2]

    m_tf = ((total/2) - (male_acc/100 * total/2))/total/2
    
    f_tf = ((total/2) - (fem_acc/100 * total/2))/total/2
    
    dpd_gb.append(abs(m_tf-f_tf))
    
    j+=1
print(dpd_gb)
j=0
for i in results:
    total = totals[j]
    
    male_t = male_tots[j]
    
    f_tot = total - male_t

    fem_acc = i[1][1][3]
    
    male_acc = i[1][1][2]

    m_tf = ((male_t) - (male_acc/100 * male_t))/male_t
    
    f_tf = ((f_tot) - (fem_acc/100 * f_tot))/f_tot
    
    dpd_or.append(abs(m_tf-f_tf))
    
    j+=1
print(dpd_or)

#now plot this 

x = model_names

plt.plot(x,dpd_gb, label = 'GB DPD scores',linestyle = '--')
plt.scatter(x,dpd_gb)

plt.plot(x,dpd_or, label = 'Base DPD scores',linestyle = '--')
plt.scatter(x,dpd_or)

plt.xlabel('Model')

plt.ylabel('DPD score')

plt.title('A comparison of GB and Base model DPDs')

plt.legend()

#

# Indexing results[Test set][model][test_set][metric]

# Loops to get single vals. 
# model = [0 - GB, 1 - Neutral, 2- Female, 3- male]
# Test_set = [0- neutral, 1- GB, 2 - female, 3- male]
# metric = [0 - Ovell accuracy, 1 - F-score, 2 - male acc, 3 - female acc]

#print(results[0][0][0][1])

print(results[6][0][0][3])
```
